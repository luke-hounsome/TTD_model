## Combine datasets of all countries


```python
countries = []
data_dir = os.path.join(
    "..", "models", "simulated_data", "230315_disease_severity_scan"
)

for file in os.listdir(data_dir):
    if file[0] != ".":  # avoid hidden files
        country = file[:-4]  # assumes country.csv
        countries.append(country)

t_dfs = []
print(countries)
for country in countries:
    df = pd.read_csv(os.path.join(data_dir, country + ".csv"))
    df["pop"] = df.apply(
        lambda row: list(map(int, row["y1_0"].strip("[, ]").split(",")))[0], axis=1
    )
    df["pop/n"] = df.apply(lambda row: row["pop"] / row["n"], axis=1)

    t_df = df[["R0", "ihr", "seed_t", "pop", "n", "pop/n"]]
    t_dfs.append(t_df)

df_merged = pd.concat(t_dfs, ignore_index=True, sort=False)
```

```python
ratios = np.sort(df_merged["pop/n"].unique())

quantiles = []
print(ratios)

for ratio in ratios:
    t_seed = df_merged.loc[np.isclose(df_merged["pop/n"], ratio)]["seed_t"]
    quantiles.append(np.quantile(t_seed, q=[0.05, 0.25, 0.5, 0.75, 0.95]))


quantiles = np.array(quantiles)
print(quantiles)


plt.xlabel("log(proportion)")
plt.ylabel("First case in UK (days)")
plt.fill_between(
    1 / ratios,
    quantiles[:, 0],
    quantiles[:, 4],
    color="teal",
    alpha=0.2,
    label="0.5 - 0.95",
)
plt.fill_between(
    1 / ratios,
    quantiles[:, 1],
    quantiles[:, 3],
    color="teal",
    alpha=0.5,
    label="0.25 - 0.75",
)
plt.ylim(bottom=0)
plt.semilogx(1 / ratios, quantiles[:, 2], "teal", label="Median")
plt.legend()
plt.savefig("ratio.png", dpi=300)
```

## Prepare data

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.tree import DecisionTreeRegressor

X = df_merged[["pop/n", "R0"]]
y = df_merged[["seed_t"]]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


X_scaler = preprocessing.StandardScaler().fit(X_train)
y_scaler = preprocessing.StandardScaler().fit(y_train)

X_train = X_scaler.transform(X_train)
X_test = X_scaler.transform(X_test)

y_train = y_scaler.transform(y_train)
y_test = y_scaler.transform(y_test)
```

## Linear regression

Linear regression does not perform well using pop/n instead of ppo and n

```python
reg = LinearRegression().fit(X_train, y_train)
reg.score(X_test, y_test)
```

```python
y_pred = reg.predict(X_test)
plt.figure(figsize=(16, 8))
plt.scatter(y_test, y_pred, marker="|")
plt.xlabel("test seed time")
plt.ylabel("predicted seed time")
```

```python
reg.coef_
```

## Decision tree regression

Decision tree performance is similar using pop/n or pop and n 

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
```

```python
leaf_sizes = [5000, 10000, 15000, 17500, 20000]
cv_scores = []

for leaf_size in leaf_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    reg = DecisionTreeRegressor(min_samples_leaf=leaf_size)
    scores = cross_val_score(reg, X_train, y_train, cv=5)
    cv_scores.append(np.mean(scores))
print(cv_scores)
plt.plot(leaf_sizes, cv_scores)
plt.xlabel("leaf_size")
plt.ylabel("cross validation score")
```

```python
from sklearn.metrics import mean_absolute_error

reg = DecisionTreeRegressor(min_samples_leaf=15000)


reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
print(y_test.shape)


print("Testing score:", reg.score(X_test, y_test))
print("MAE:", mean_absolute_error(y_pred, y_test))


print("d")
plt.figure(figsize=(16, 8))
plt.scatter(y_test, y_pred, marker="|")
plt.plot([0, 220], [0, 220])

plt.xlabel("test seed time")
plt.ylabel("predicted seed time")
```

## Make this plot better

```python
df_av = df_merged.groupby(["pop/n", "R0"])["seed_t"].mean().reset_index()

X = df_av[["pop/n", "R0"]]
y = df_av[["seed_t"]]

y_pred = reg.predict(X)
plt.figure(figsize=(16, 8))
plt.scatter(y, y_pred, marker="|")
plt.plot([0, 220], [0, 220])

plt.xlabel("test seed time")
plt.ylabel("predicted seed time")
```

```python
# plot tree
from sklearn import tree

print(reg.get_depth())
print(reg.get_n_leaves())
print(reg.get_params())
plt.figure(figsize=(32, 16))
tree.plot_tree(reg, feature_names=["pop/n", "R0"], filled=True)
```

## Do minimal cost complexity pruning to get interpretible tree

```python
path = reg.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
regs = []
cv_scores = []
for ccp_alpha in ccp_alphas:
    reg = DecisionTreeRegressor(ccp_alpha=ccp_alpha)
    scores = cross_val_score(reg, X_train, y_train, cv=5)
    cv_scores.append(np.mean(scores))
    reg.fit(X_train, y_train)
    regs.append(reg)
```

```python
node_counts = [reg.tree_.node_count for reg in regs]
depth = [reg.tree_.max_depth for reg in regs]
fig, ax = plt.subplots(4, 1, figsize=(8, 16))
ax[0].plot(ccp_alphas[:-1], impurities[:-1], marker="o", drawstyle="steps-post")
ax[0].set_xlabel("alpha")
ax[0].set_ylabel("total impurity of leaves")
ax[0].set_title("Total Impurity vs effective alpha for training set")
ax[1].plot(ccp_alphas, node_counts, marker="o", drawstyle="steps-post")
ax[1].set_xlabel("alpha")
ax[1].set_ylabel("number of nodes")
ax[1].set_title("Number of nodes vs alpha")
ax[2].plot(ccp_alphas, depth, marker="o", drawstyle="steps-post")
ax[2].set_xlabel("alpha")
ax[2].set_ylabel("depth of tree")
ax[2].set_title("Depth vs alpha")
ax[3].plot(ccp_alphas, cv_scores, marker="o", drawstyle="steps-post")
ax[3].set_xlabel("alpha")
ax[3].set_ylabel("CV score")
ax[3].set_title("CV score vs alpha")
fig.tight_layout()
```

```python
reg = DecisionTreeRegressor(ccp_alpha=ccp_alphas[-4])
scores = cross_val_score(reg, X_train, y_train, cv=5)
print(np.mean(scores))
reg.fit(X_train, y_train)
print("Testing score:", reg.score(X_test, y_test))
plt.figure(figsize=(32, 16))
tree.plot_tree(reg, feature_names=["pop/n", "R0"], filled=True)
```
